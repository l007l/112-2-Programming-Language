```python
import json
import pandas as pd
```

```python
f = open('C:/Users/user/PL/HW2/tip.json')
data = json.load(f)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#以直方圖顯示光顧餐廳的顧客性別
sex_counts = df['sex'].value_counts()

plt.figure(figsize=(4, 6))
plt.bar(sex_counts.index, sex_counts.values)
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
#以直方圖顯示光顧餐廳的顧客是否吸煙者
smoker_counts = df['smoker'].value_counts()

plt.figure(figsize=(4, 6))
plt.bar(smoker_counts.index, smoker_counts.values)
plt.title('Smoker Distribution')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()
```

```python
#以圓餅圖顯示顧客光顧餐廳的日子各佔百分比
day_counts = df['day'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(day_counts, labels=day_counts.index, autopct='%.1f%%')
plt.title('Day Distribution')
plt.axis('equal')
plt.show()
```

```python
#以直方圖顯示顧客光顧餐廳的時間
time_counts = df['time'].value_counts()

plt.figure(figsize=(4, 6))
plt.bar(time_counts.index, time_counts.values)
plt.title('Time Distribution')
plt.xlabel('time')
plt.ylabel('Count')
plt.show()
```

```python
#以圓餅圖顯示光顧餐廳的顧客人數各佔百分比
size_counts = df['size'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(size_counts, labels=size_counts.index, autopct='%.1f%%')
plt.title('Size Distribution')
plt.axis('equal')
plt.show()
```

```python
#創建圖表大小（尺寸為英吋，這裡的 10 和 6 分別代表圖表的寬度和高度）
plt.figure(figsize=(10, 6))
#繪製散點圖，利用alpha指定散點圖中點的透明度，方便觀察數據密度/重疊情況（相同數據出現次數越多，顏色越深）
plt.scatter(df['total_bill'], df['tip'], alpha=0.5)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill (USD)')
plt.ylabel('Tip (USD)')
#取得當前圖表的軸
ax = plt.gca()
#設定x和y軸的刻度間隔，方便閱讀散點圖分佈位置
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
#顯示網格，進行對位，方便閱讀
plt.grid(True)
plt.show()
```
