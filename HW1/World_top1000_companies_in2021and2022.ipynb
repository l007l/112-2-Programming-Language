```python
import pandas as pd

data2021 = pd.read_csv('C:/Users/user/PL/HW1/fortune1000_companies_2021.csv')
data2022 = pd.read_csv('C:/Users/user/PL/HW1/fortune1000_companies_2022.csv')
```

```python
# 欄位對應位置的數據轉成set
campanies2021 = set(data2021.iloc[:, 2])
campanies2022 = set(data2022.iloc[:, 1])
```

```python
# 檢查類型是否為set
type(campanies2021)
```

```python
# 檢查類型是否為set
type(campanies2022)
```

```python
# 計算差集、交集和聯集並順序排列
difference2021_set = sorted(campanies2021.difference(campanies2022))
difference2022_set = sorted(campanies2022.difference(campanies2021))
intersection_set = sorted(campanies2021.intersection(campanies2022))
union_set = sorted(campanies2021.union(campanies2022))
```

```python
# 計算差集、交集和聯集結果的筆數
no_difference2021_set = len(difference2021_set)
no_difference2022_set = len(difference2022_set)
no_intersection_set = len(intersection_set)
no_union_set = len(union_set)
```

```python
# 將 DataFrame 轉換為字典，其中鍵為 Name 欄位的值，值為 Rank 欄位的值
rank2021 = data2021.set_index('Name')['Rank'].to_dict()
rank2022 = data2022.set_index('Name')['rank in 2022'].to_dict()
```

```python
# 迴圈執行差集結果中公司名稱和對應排名
print("差集，2021年度有進世界公司財富榜頭1000名但沒進2022年度的公司有：\n")
for company_name in difference2021_set:
    if company_name in rank2021:
        rank = rank2021[company_name]
        print(f"Rank: {rank}, Company: {company_name}")
print("\n筆數：", no_difference2021_set)
```

```python
print("差集，2022年度有進世界公司財富榜頭1000名但沒進2021年度的公司有：\n")
for company_name in difference2022_set:
    if company_name in rank2022:
        rank = rank2022[company_name]
        print(f"Rank: {rank}, Company: {company_name}")
print("\n筆數：", no_difference2022_set)
```

```python
print("交集，2021及2022年度均有進世界公司財富榜頭1000名的公司有：\n")
for item in intersection_set:
    print(item)
print("\n筆數：", no_intersection_set)
```

```python
print("聯集，2021或2022年度有進世界公司財富榜頭1000名的公司有：\n")
for item in union_set:
    print(item)
print("\n筆數：", no_union_set)
```
