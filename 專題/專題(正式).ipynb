```python
current_url = driver.current_url
response = requests.get(current_url)
job=[]
if response.status_code == 200:
    # 解析 HTML 內容
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 找到所有 class 為 'job__title' 的元素
    job_title_elements = soup.find_all('a', class_='job__title')
    
    # 提取並打印每個職業名稱
    for job_title_element in job_title_elements:
        job_title = job_title_element.get_text(strip=True)
        print('職業名稱:', job_title)
        job.append(job_title)
else:
    print('無法請求網頁，狀態碼:', response.status_code)
```

```python
company=[]
if response.status_code == 200:
    # 解析 HTML 內容
    soup = BeautifulSoup(response.text, 'html.parser')
    # 找到所有 class 為 'job__title' 的元素
    job__comp__name_elements = soup.find_all('span', class_='job__comp__name')
    
    # 提取並打印每個職業名稱
    for job__comp__name_element in job__comp__name_elements:
        job__comp__name = job__comp__name_element.get_text(strip=True)
        print('公司名稱:', job__comp__name)
        company.append(job__comp__name)
else:
    print('無法請求網頁，狀態碼:', response.status_code)
```

```python
salery=[]
if response.status_code == 200:
    # 解析 HTML 內容
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 找到所有 class 為 'job__title' 的元素
    salery_elements = soup.find_all('p', class_='job__salary')
    
    # 提取並打印每個職業名稱
    for salery_element in salery_elements:
        job_salery = salery_element.get_text(strip=True)
        print('薪水:',job_salery)
        salery.append(job_salery.replace(',',''))
else:
    print('無法請求網頁，狀態碼:', response.status_code)
    
import re

def extract_salary_range(text):
    # 使用正則表達式提取數字
    numbers = re.findall(r'\d+', text)
    
    # 如果找到兩個數字,則認為是一個區間
    if len(numbers) == 2:
        min_salary = int(numbers[0])
        max_salary = int(numbers[1])
        return f"{min_salary} - {max_salary}"
    # 如果只找到一個數字,則認為是一個固定值
    elif len(numbers) == 1:
        salary = int(numbers[0])
        return str(salary)
    else:
        return "無法提取薪資資訊"

# 定義一個列表，包含需要處理的文本
texts = salery

# 使用列表推導將列表中的每個元素都傳遞給 extract_salary_range 函數
salary_ranges = [extract_salary_range(text) for text in texts]

# 顯示結果
print(salary_ranges)
```

```python
des=[]
if response.status_code == 200:
    # 解析 HTML 內容
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 找到所有 class 為 'job__title' 的元素
    des_elements = soup.find_all('p', class_='job__intro')
    
    # 提取並打印每個職業名稱
    for des_element in des_elements:
        job_des = des_element.get_text(strip=True)
        print('職業名稱:', job_des)
        des.append(job_des.replace('/n',''))
else:
    print('無法請求網頁，狀態碼:', response.status_code)
```

```python
a=[]
if response.status_code == 200:
    # 解析 HTML 內容
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 找到所有 class 為 'job__title' 的元素
    job_place_elements = soup.find_all('ul', class_='job__summaries')
    
    # 提取並打印每個職業名稱
    for job_place_element in job_place_elements:
        job_place = job_place_element.get_text(strip=True)
        print(job_place)
        a.append(job_place)
        
else:
    print('無法請求網頁，狀態碼:', response.status_code)

    
place = [item[:7] for item in a]
print(place)
```

```python
list_combined = list(zip(job,company,salary_ranges,des,place))
def clean_string(s):
    return s.replace('\n', '').replace('\r', '').strip()

# 清理列表中的每一个元素
cleaned_list_combined = []
for i in list_combined:
    cleaned_job = tuple(clean_string(field) for field in i)
    cleaned_list_combined.append(cleaned_job)

# 显示清理后的列表
for i in cleaned_list_combined:
    print(i)
```

```python

# Load the CSV file
df = pd.read_csv('Job.csv')

# Define the segmentation function
def seg_words(text):
    return " ".join(jieba.cut(text))

# Apply the segmentation function to the text column
df['cut'] = df['des'].apply(seg_words)
a=[]
# Define the Gradio interface
def job_search(keyword, job_place, salary):
    a = []  # Initialize the list to store results
    
    for index, row in df.iterrows():
        if row['salary_ranges'] is not None:
            numbers = re.findall(r'\d+', str(row['salary_ranges']))
            if len(numbers) == 1:
                if int(salary) >= int(numbers[0]) and keyword in row['cut'] and job_place in row['place']:
                    for col in ['job', 'company', 'salary_ranges', 'place', 'des']:
                        a.append(col + " ： " + str(row[col]))
            if len(numbers) == 2:
                if int(numbers[0]) <= int(salary) <= int(numbers[1]) and keyword in row['cut'] and job_place in row['place']:
                    for col in ['job', 'company', 'salary_ranges', 'place', 'des']:
                        a.append(col + " ： " + str(row[col]))
    
    output = '\n'.join(a)
    return output

# Create the Gradio interface
gr.Interface(
    fn=job_search,
    inputs=["text", "text", "number"],
    outputs="text",
    title="Job Search",
    description="輸入職缺關鍵字、職缺地點、預期薪水"
).launch()
```
