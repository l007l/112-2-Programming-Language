```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('D:/Google/adult.csv')
```

```python
average_income = np.mean(data['Income'])
std_income = np.std(data['Income'])
median_income = np.median(data['Income'])
```

```python
Income = data['Income']
```

```python
plt.hist(Income, bins=30, edgecolor='black')
plt.title('Incomes Distribution')
plt.xlabel('Income')
plt.ylabel('No.of_people')

plt.axvline(x=average_income, color='red', linestyle='--')
plt.axvline(x=std_income, color='green', linestyle='--')
plt.axvline(x=median_income, color='blue', linestyle='--')

plt.text(1100000,8300,'--average_income', color='red')
plt.text(1100000,7800,'--std_income', color='green')
plt.text(1100000,7300,'--median_income', color='blue')
plt.show()
```
